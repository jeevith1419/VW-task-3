```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

pd.set_option("display.max_columns",None)
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]

df = pd.read_csv(filename)

df.head()
```

```python
print("Dataset Shape :", df.shape)

print("\nColumns")

print(df.columns)

print("\nInformation")

df.info()

print("\nSummary")

display(df.describe(include="all"))
```

```python
print(df.isnull().sum())
```

```python
print("Duplicate Rows :",df.duplicated().sum())

df=df.drop_duplicates()

print("New Shape :",df.shape)
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="Revenue",data=df)

plt.title("Revenue Distribution")

plt.show()

print(df["Revenue"].value_counts())
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["BounceRates"],bins=30)

plt.title("Bounce Rate Distribution")

plt.xlabel("Bounce Rate")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["ExitRates"],bins=30,color="orange")

plt.title("Exit Rate Distribution")

plt.xlabel("Exit Rate")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["ProductRelated"],bins=30)

plt.title("Product Related Pages Visited")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["ProductRelated"],bins=30)

plt.title("Product Related Pages Visited")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(x="VisitorType",data=df)

plt.title("Visitor Type")

plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(x="TrafficType",data=df)

plt.title("Traffic Source")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(x="Month",data=df,
order=["Feb","Mar","May","June","Jul","Aug","Sep","Oct","Nov","Dec"])

plt.title("Monthly Website Traffic")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="Weekend",hue="Revenue",data=df)

plt.title("Weekend vs Revenue")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="Revenue",y="BounceRates",data=df)

plt.title("Bounce Rate vs Revenue")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="Revenue",y="ExitRates",data=df)

plt.title("Exit Rate vs Revenue")

plt.show()
```

```python
numeric=df.select_dtypes(include=np.number)

plt.figure(figsize=(14,10))

sns.heatmap(

numeric.corr(),

annot=True,

cmap="coolwarm"

)

plt.title("Correlation Heatmap")

plt.show()
```

```python
print("="*60)
print("WEB TRAFFIC ANALYTICS REPORT")
print("="*60)
print("Total Sessions :",len(df))
print()
print("Average Bounce Rate :",round(df["BounceRates"].mean(),3))
print()
print("Average Exit Rate :",round(df["ExitRates"].mean(),3))
print()
df["TotalDuration"] = df["Administrative_Duration"] + df["Informational_Duration"] + df["ProductRelated_Duration"]
print("Average Session Duration :",round(df["TotalDuration"].mean(),2),"seconds")
print()
print("Average Product Pages :",round(df["ProductRelated"].mean(),2))
print()
print("Most Common Visitor Type :",df["VisitorType"].mode()[0])
print()
print("Most Common Traffic Type :",df["TrafficType"].mode()[0])
print()
print("Revenue Conversion Rate :",round(df["Revenue"].mean()*100,2),"%")
print("="*60)
```

```python
print("BUSINESS INSIGHTS")

print()

print("1. Higher Bounce Rates generally reduce conversions.")

print("2. Higher Exit Rates indicate visitors leave without purchasing.")

print("3. Longer session durations improve purchase probability.")

print("4. Returning visitors contribute significantly to revenue.")

print("5. Product-related page visits strongly influence purchases.")

print("6. Certain traffic sources generate better conversions.")
```
